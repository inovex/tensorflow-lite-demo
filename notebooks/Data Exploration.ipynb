```python
imdir = "../data/images"
```

```python
gen = ImageDataGenerator(dtype=np.uint8, rescale=1./255, validation_split=0.2)
train = gen.flow_from_directory(imdir, batch_size=32, target_size=(224, 224), color_mode='rgb', class_mode="categorical", subset="training")
val = gen.flow_from_directory(imdir, batch_size=32, target_size=(224, 224), color_mode='rgb', class_mode="categorical", subset="validation")
```
